```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import glob
import os

path_SPARC = "my_data/SPARC/" # use your path
path_TGFBI = "my_data/TGFBI/" # use your path
path_control = "my_data/control/" # use your path
all_files_SPARC = glob.glob(os.path.join(path_SPARC , "*.csv"))
all_files_TGFBI = glob.glob(os.path.join(path_TGFBI , "*.csv"))
all_files_control = glob.glob(os.path.join(path_control , "*.csv"))

li_SPARC = []
li_TGFBI = []
li_control = []

for filename in all_files_SPARC:
    df_SPARC = pd.read_csv(filename, index_col=None, header=0,skiprows=17)
    li_SPARC.append(df_SPARC)

for filename in all_files_TGFBI:
    df_TGFBI = pd.read_csv(filename, index_col=None, header=0,skiprows=17)
    li_TGFBI.append(df_TGFBI)

for filename in all_files_control:
    df_control = pd.read_csv(filename, index_col=None, header=0,skiprows=17)
    li_control.append(df_control)

frame_SPARC = pd.concat(li_SPARC, axis=0, ignore_index=True)
frame_TGFBI = pd.concat(li_TGFBI, axis=0, ignore_index=True)
frame_control = pd.concat(li_control, axis=0, ignore_index=True)
```

```python
means_SPARC = frame_SPARC.groupby('x').mean()
means_SPARC
```

```python
means_TGFBI = frame_TGFBI.groupby('x').mean()
means_TGFBI
```

```python
means_control = frame_control.groupby('x').mean()
means_control
```

```python
plt.rc('font', family='Times New Roman')
plt.rc('axes', labelsize=12, labelpad=13) 
plt.rc('xtick', labelsize=12) 
plt.rc('ytick', labelsize=12) 

fig, ax = plt.subplots()

ax = means_SPARC.plot(ax=ax, x = 'x.1', y = 'y', kind='line', color="#52b69a", label="Stem Cells (SPARC-influenced)", linewidth=2)
ax = means_SPARC.plot(ax=ax, x = 'x.1', y = 'y.1', kind='line', color="#99d98c", label="Non-Stem Cells (SPARC-influenced)", linewidth=2)
ax = means_TGFBI.plot(ax=ax, x = 'x.1', y = 'y', kind='line', color="#168aad", label="Stem Cells (TGFBI-influenced)", linewidth=2)
ax = means_TGFBI.plot(ax=ax, x = 'x.1', y = 'y.1', kind='line', color="#184e77", label="Non-Stem Cells (TGFBI-influenced)", linewidth=2)
ax = means_control.plot(ax=ax, x = 'x.1', y = 'y', kind='line', color="#ee6055", label="Stem Cells (Control)", linewidth=2)
ax = means_control.plot(ax=ax, x = 'x.1', y = 'y.1', kind='line', color="#bc4b51", label="Non-Stem Cells (Control)", linewidth=2)

ax.set_xlabel("Time (timesteps)")
ax.set_ylabel("Number of Cells (cells)")
ax.spines['top'].set_visible(True)
ax.spines['right'].set_visible(True)
ax.spines['bottom'].set_visible(True)
ax.spines['left'].set_visible(True)
ax.tick_params(direction="in")
plt.ylim([0, 800])
plt.xlim([0, 1500])

pos = ax.get_position()
ax.set_position([pos.x0, pos.y0, pos.width, pos.height * 0.85])
plt.rcParams["figure.figsize"] = (20,20)
plt.title("Average Number of Stem and Non-Stem Cells Versus Time for SPARC-influenced, TGFBI-influenced, and control group", fontsize = 30)
plt.grid()

plt.show()
```

```python
import glob
import os

path_CSNCs = "my_data2/CSNCs/" # use your path
path_CSCs = "my_data2/CSCs/" # use your path
all_files_CSNCs = glob.glob(os.path.join(path_CSNCs , "*.csv"))
all_files_CSCs = glob.glob(os.path.join(path_CSCs , "*.csv"))

li_CSNCs = []
li_CSCs = []

for filename in all_files_CSNCs:
    df_CSNCs = pd.read_csv(filename, index_col=None, header=0,skiprows=19)
    li_CSNCs.append(df_CSNCs)

for filename in all_files_CSCs:
    df_CSCs = pd.read_csv(filename, index_col=None, header=0,skiprows=19)
    li_CSCs.append(df_CSCs)

frame_CSNCs = pd.concat(li_CSNCs, axis=0, ignore_index=True)
frame_CSCs = pd.concat(li_CSCs, axis=0, ignore_index=True)
```

```python
means_CSNCs = frame_CSNCs.groupby('x').mean()
means_CSNCs
```

```python
means_CSCs = frame_CSCs.groupby('x').mean()
means_CSCs
```

```python
means_CSNCs = means_CSNCs.drop(['x.1', 'color', 'pen down?', 'color.1', 'pen down?.1', 'x.2', 'color.2', 'pen down?.2', 'x.3', 'color.3', 'pen down?.3'], axis=1)
```
